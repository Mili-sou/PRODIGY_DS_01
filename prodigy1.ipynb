# Plotting Bar Chart

```python
#ploting of countries to their respective population in 2022.
plot_countries = ['Australia', 'Belgium', 'Brazil', 'Afghanistan', 'Colombia', 'France']
#filtering the data based on the countries to be ploted.
data = df[df['Country Name'].isin(plot_countries)]

#Plotting of bar chart
plt.bar (data['Country Name'], data['2020'],color="pink")
plt.xlabel ('Country Name')
plt.ylabel ('Population in 2020')
plt.title('Population of countries in 2020')
plt.show()
```

# Plotting Histogram

```python
#plotting histogram of a particular year to visualize the density of population.
plt.hist(data['2020'],bins=10,color="red")
plt.xlabel ('Year')
plt.ylabel ('Density of population')
```
